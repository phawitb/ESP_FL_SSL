```python
import os
import numpy as np
import pandas as pd
import random
```

```python
# === CONFIG ===
DATASET_PATH = "UCI_HAR_Dataset"
OUTPUT_DIR = "simulated_sensor_data"
LABEL_PERCENTAGE = 0.2  # à¸ªà¸±à¸”à¸ªà¹ˆà¸§à¸™à¸‚à¸­à¸‡à¹„à¸Ÿà¸¥à¹Œà¸—à¸µà¹ˆà¸¡à¸µ label
WINDOW_SIZE = 2000
NUM_SIGNALS = 9
```

```python
def load_raw_signals_df():
    folder = os.path.join(DATASET_PATH, "train", "Inertial Signals")
    signal_files = [
        "body_acc_x_train.txt", "body_acc_y_train.txt", "body_acc_z_train.txt",
        "body_gyro_x_train.txt", "body_gyro_y_train.txt", "body_gyro_z_train.txt",
        "total_acc_x_train.txt", "total_acc_y_train.txt", "total_acc_z_train.txt"
    ]
    signal_names = [
        'acc_x', 'acc_y', 'acc_z',
        'gyro_x', 'gyro_y', 'gyro_z',
        'acc_total_x', 'acc_total_y', 'acc_total_z'
    ]
    
    signals = []
    for filename in signal_files:
        path = os.path.join(folder, filename)
        print(f"ðŸ“¥ Loading {filename}")
        data = np.loadtxt(path)  # Shape: (7352, 128)
        signals.append(data)
    
    # Stack into shape: (7352, 128, 9)
    X = np.stack(signals, axis=-1)
    num_windows, window_size, num_signals = X.shape

    # Flatten to shape: (7352 * 128, 9)
    X_flat = X.reshape(-1, num_signals)

    # Load labels and repeat each label for 128 timesteps
    y_path = os.path.join(DATASET_PATH, "train", "y_train.txt")
    y = np.loadtxt(y_path).astype(int)
    y_flat = np.repeat(y, window_size)  # Shape: (7352 * 128,)

    # Create DataFrame
    df = pd.DataFrame(X_flat, columns=signal_names)
    df['label'] = y_flat

    return df
```

```python
df = load_raw_signals_df()
df
```

```python
df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)
df_shuffled
```

```python
def export_windows_from_df(df, label_percentage, window_size=128):
    os.makedirs(OUTPUT_DIR, exist_ok=True)

    num_rows = len(df)
    num_windows = num_rows // window_size
    for i in range(num_windows):
        start_idx = i * window_size
        end_idx = start_idx + window_size
        window_df = df.iloc[start_idx:end_idx].copy()

        if len(window_df) < window_size:
            continue  # skip incomplete windows

        is_labeled = random.random() < label_percentage
        filename = f"window_{i:04d}_{'labeled' if is_labeled else 'unlabeled'}.csv"
        filepath = os.path.join(OUTPUT_DIR, filename)

        if not is_labeled:
            window_df = window_df.drop(columns=['label'], errors='ignore')

        window_df.to_csv(filepath, index=False)
        print(f"âœ… Saved: {filepath}")
```

```python
label_percentage = 0.4
export_windows_from_df(df_shuffled, label_percentage, window_size=WINDOW_SIZE)
```

```python




# === STEP 1: Load raw inertial signals ===
def load_raw_signals():
    folder = os.path.join(DATASET_PATH, "train", "Inertial Signals")
    signal_files = [
        "body_acc_x_train.txt", "body_acc_y_train.txt", "body_acc_z_train.txt",
        "body_gyro_x_train.txt", "body_gyro_y_train.txt", "body_gyro_z_train.txt",
        "total_acc_x_train.txt", "total_acc_y_train.txt", "total_acc_z_train.txt"
    ]
    signals = []
    for filename in signal_files:
        path = os.path.join(folder, filename)
        print(f"ðŸ“¥ Loading {filename}")
        data = np.loadtxt(path)  # Shape: (7352, 128)
        signals.append(data)
    # Stack along last axis â†’ shape: (7352, 128, 9)
    X = np.stack(signals, axis=-1)
    y_path = os.path.join(DATASET_PATH, "train", "y_train.txt")
    y = np.loadtxt(y_path).astype(int)
    return X, y

# === STEP 2: Export each window to CSV ===
def export_windows(X, y, label_percentage):
    os.makedirs(OUTPUT_DIR, exist_ok=True)

    for i, (window, label) in enumerate(zip(X, y)):
        is_labeled = random.random() < label_percentage
        filename = f"window_{i:04d}_{'labeled' if is_labeled else 'unlabeled'}.csv"
        filepath = os.path.join(OUTPUT_DIR, filename)

        df = pd.DataFrame(window, columns=[
            'acc_x', 'acc_y', 'acc_z',
            'gyro_x', 'gyro_y', 'gyro_z',
            'acc_total_x', 'acc_total_y', 'acc_total_z'
        ])
        if is_labeled:
            df['label'] = label
        df.to_csv(filepath, index=False)
        print(f"âœ… Saved: {filepath}")

# === MAIN ===
if __name__ == "__main__":
    print("ðŸš€ Loading raw UCI HAR signals...")
    X, y = load_raw_signals()
    print("ðŸ“¦ Exporting sensor simulation files to CSV...")
    export_windows(X, y, LABEL_PERCENTAGE)
    print("ðŸŽ‰ Done! Files are ready in:", OUTPUT_DIR)
```
